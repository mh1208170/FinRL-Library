```python
import pandas as pd
import pyEX as p
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))


df = pd.read_csv("gc-1m.csv", sep=";",nrows=1000)
df.columns=["Date","Time", "Open", "High", "Low", "Close", "Volume"]
```

```python
dt=pd.to_datetime(df['Date'] + ' ' + df['Time'])
df["DataTime"]=dt
df.drop(columns=["Time","Date","Volume"],inplace=True)
df.set_index(["DataTime"],inplace=True)
#print(df.head())
exp1 = df["Close"].ewm(span=12, adjust=False).mean()
exp2 = df["Close"].ewm(span=26, adjust=False).mean()
macd = exp1-exp2
exp3 = macd.ewm(span=9, adjust=False).mean()
df["exp12"]=exp1
df["exp26"]=exp2
df["macd"]=exp3
#print(df.head())
df["resutls"]=df.Close.diff(periods=-9)
#print(df.head(10))
df['action']=0
df["action"][(df["resutls"] > 0.5)]= 1
df["action"][(df["resutls"] < -0.5)]= 2
#>>> df["D"][(df["A"] == 0) & ((df["B"] > 0) | (df["C"] > 0))] = "c2"
print(df.tail(50))
df.to_csv('gold.csv')
```
